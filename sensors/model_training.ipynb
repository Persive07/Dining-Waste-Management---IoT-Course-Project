```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.preprocessing import MinMaxScaler
import joblib

df = pd.read_csv('/content/mess_10s_kg_data.csv')
X_raw = df.iloc[:, 0:10].values 
y_raw = df.iloc[:, 11:].values 

scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X_raw)
y_scaled = scaler_y.fit_transform(y_raw)

def create_lstm_sequences(X, y, time_steps=6):
    xs, ys = [], []
    for i in range(len(X) - time_steps):
        xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(xs), np.array(ys)

X_seq, y_seq = create_lstm_sequences(X_scaled, y_scaled)

model = models.Sequential([
    layers.LSTM(64, activation='tanh', input_shape=(6, 10), return_sequences=True),
    layers.LSTM(32, activation='tanh'),
    layers.Dense(32, activation='relu'),
    layers.Dense(4, activation='linear') 
])
def asymmetric_mse(y_true, y_pred):
    error = y_true - y_pred
    return tf.reduce_mean(tf.where(error > 0, tf.square(error) * 2.0, tf.square(error)))

model.compile(optimizer='adam', loss=asymmetric_mse)
print("Training LSTM Brain...")
model.fit(X_seq, y_seq, epochs=30, batch_size=32, verbose=1)

model.save('lstm_mess_brain.h5')
joblib.dump(scaler_x, 'scaler_x.pkl')
joblib.dump(scaler_y, 'scaler_y.pkl')
print("LSTM Model Ready.")
```
